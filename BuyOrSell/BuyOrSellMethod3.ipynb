# 3. MACD Crosses Signal Line
https://www.udemy.com/course/algorithmic-trading-quantitative-analysis-using-python/learn/lecture/14650722#content

The strategy is to buy if the MACD cuts the Signal line from below => bullisch and vice versa

 
Notes:
- Depends on the choices of the length of the moving averages
- The course talks about shorter periods(not days but minutes), but I wanted to compare using the same playing field

## 3.1 methods for everyone

```python
# from 2.1
def MACD(DF,short, long, signal):# 12,26, 9 are standards
    df=DF.copy()
    df["MAFast"]=df["Adj Close"].ewm(span=short, min_periods=short).mean()
    df["MASlow"]=df["Adj Close"].ewm(span=long, min_periods=long).mean()
    df["MACD"]=df["MAFast"]-df["MASlow"]
    df["SignalLine"]=df["MACD"].ewm(span=signal, min_periods=signal).mean()
    df.dropna(inplace=True)
    return df
```

```python
# Parameters
Short=12
Long=26
Signal=9
```

```python
def MACDSignalCross(df):
    start=26
    buy=[]
    sell=[]
    # same as above(technically useless)
    short=[]
    BuyBack=[]
    for _ in range(start):
        buy.append(np.nan)
        sell.append(np.nan)
        short.append(np.nan)
        BuyBack.append(np.nan)
    above=0 # 1 above -1 below
    for i in range(start,df.shape[0]):
        # above 0
        if (above!=1 and (df["MACD"][i]>df["SignalLine"][i])):
            buy.append(df["Adj Close"][i])
            sell.append(np.nan)
            BuyBack.append(df["Adj Close"][i])
            short.append(np.nan)
            above=1
                
        # below 0  
        elif (above!=-1 and (df["MACD"][i]<df["SignalLine"][i])):
            buy.append(np.nan)
            sell.append(df["Adj Close"][i])
            short.append(df["Adj Close"][i])
            BuyBack.append(np.nan)
            above=-1
            
        #do nothing
        else:
            sell.append(np.nan)
            buy.append(np.nan)
            short.append(np.nan)
            BuyBack.append(np.nan)
            
    return buy, sell, short, BuyBack
```

```python
def createLog(df):# creating a shortened log of the price when we bought/sold...
    bought=[df["Adj Close"][x] for x in df.index if not math.isnan(df["Buy"][x])]
    sold=[df["Adj Close"][x] for x in df.index if not math.isnan(df["Sell"][x])]
    shorted=[df["Adj Close"][x] for x in df.index if not math.isnan(df["Short"][x])]
    boughtback=[df["Adj Close"][x] for x in df.index if not math.isnan(df["BuyBack"][x])]
    return bought, sold, shorted, boughtback
# could have just used the lists from above this is just for demonstration
```
